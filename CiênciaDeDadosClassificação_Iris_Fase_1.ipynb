```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
#Função que compara cada linha do dataframe com os resultados.
def compara(df):
    if df['Predito']==df['Real']:
        return True
    else:
        return False
#Lista com os possíveis nomes das classes.
nomes_target = datasets.load_iris().target_names
#Função que define os nomes das classes no dataframe de resultados.
#Se no dataframe tiver o valor 0, a função vai retornar nomes_target[0].
def define_nomes(valor):
    return nomes_target[int(valor)]
```

```python
#Faz o download da base de dados. Coloca as variáeis independentes em X e a variável dependente em y.
X, y = datasets.load_iris(return_X_y=True, as_frame=True)
#Seleciona apenas as variáveis das pétalas.
X = X[['petal length (cm)','petal width (cm)']]
#X = X[['petal width (cm)']]
#Separa a base em treino e teste.
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
#Cria a variável que vai construir a árvore de decisão.
clf = DecisionTreeClassifier()
#Treina o modelo
clf.fit(X_train,y_train)
#Calcula a acurácia, comparando os valores que o modelo previu (com o uso de X_test) com y_test.
display(clf.score(X_test,y_test))
```

```python
#Faz o download da base de dados. Coloca as variáeis independentes em X e a variável dependente em y.
X, y = datasets.load_iris(return_X_y=True, as_frame=True)
#Seleciona apenas as variáveis das pétalas.
X = X[['petal length (cm)','petal width (cm)']]
#Separa as bases X e y em treino e teste.
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
#Cria a variável que vai construir a árvore de decisão.
clf_petalas = DecisionTreeClassifier()
#Treina o modelo
clf_petalas.fit(X_train,y_train)
#Avalia o modelo treinado
#Calcula a acurácia, comparando os valores que o modelo previu (com o uso de X_test) com y_test.
display("Acurácia sklearn = "+str(clf_petalas.score(X_test,y_test)))
######
#Cria uma estrutura com os valores preditos pelo modelo treinado.
y_pred = clf_petalas.predict(X_test)
df_resultados_petalas = pd.DataFrame([y_pred,y_test]).T
#Refaz os índices da estrutura com os valores independentes de teste.
X_test = X_test.reset_index(drop=True)
#Junta os valores de X (variáveis independentes) com os resultados preditos e reais.
df_resultados_petalas = pd.concat([X_test,df_resultados_petalas], axis = 1)
df_resultados_petalas.columns = ['petal length (cm)','petal idth (cm)','Predito','Real']
df_resultados_petalas['Acerto'] = df_resultados_petalas.apply(compara, axis = 1)
df_resultados_petalas['Predito'] = df_resultados_petalas['Predito'].apply(define_nomes)
df_resultados_petalas['Real'] = df_resultados_petalas['Real'].astype(int)
df_resultados_petalas['Real'] = df_resultados_petalas['Real'].apply(define_nomes)
acurácia = df_resultados_petalas['Acerto'].value_counts()[True]/df_resultados_petalas['Acerto'].value_counts().sum()
display("Acurácia manual = "+str(acurácia))
```
