```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
base = pd.read_csv('breast-cancer-dataset.csv')
```

```python
base
```

```python
#1 - Apague a coluna S/N
base = base.drop(columns=['S/N'])
#2 - Faça a dummificação nas variáveis categóricas.
dummies_diagnosis_result = pd.get_dummies(base['Diagnosis Result'])
dummies_breast = pd.get_dummies(base['Breast'])
dummies_breast_quadrant = pd.get_dummies(base['Breast Quadrant'])
#3 - Separes as bases em variáveis independentes e dependentes.
variaveis_independentes = base[['Age', 'Year', 'Menopause'], dummies_breast]
variaveis_dependentes = base[['Tumor Size (cm)', 'Inv-Nodes', 'Metastasis', dummies_breast_quadrant, 'History', dummies_diagnosis_result]]
#4 - Crie a base de treino e de teste.
X_train, X_test, y_train, y_test = train_test_split(variaveis_dependentes, variaveis_independentes)
#5 - Crie o classificador
clf = DecisionTreeClassifier()

str(clf.fit(X_train, y_train))

display(clf.score(X_test, y_test))
#6 - Avalie o classificador
```
